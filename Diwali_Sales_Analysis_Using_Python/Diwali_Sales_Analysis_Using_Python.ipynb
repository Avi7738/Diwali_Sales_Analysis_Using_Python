```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Drop Unwanted Columns**

```python
df.drop(['Status','unnamed1'] , axis = 1 , inplace =True)
```

# **Handling missing data**

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

# **Removing duplicates**

```python
df.drop_duplicates(inplace=True)
```

# **Converting data types**

```python
df['Amount'] = df['Amount'].astype('int')
```

# **Visualizing a data**

### **Distribution of Gender**

Shows how many males and females are in the dataset.

Helps understand the gender balance or imbalance in the sample.

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(5, 5))
plt.bar(gender_counts.index, gender_counts.values, color=['royalblue', 'peru'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### **Distribution of Age Group**

Displays how many people fall into different age categories.

Gives an overview of the age distribution within the dataset.

```python
age_group_counts = df['Age Group'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(age_group_counts.index, age_group_counts.values, color='darkgoldenrod')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### **Total Orders by Product Category**

Counts how many orders each type of product received.

Indicates which product categories are most popular based on order volume.

```python
product_orders = df.groupby('Product_Category')['Orders'].sum()
plt.figure(figsize=(10, 6))
plt.bar(product_orders.index, product_orders.values, color='slateblue')
plt.title('Total Orders by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```

### **Total Amount by State**

Shows how much money each state contributed in sales.

Gives insight into which states generate the highest revenue.

```python
state_amount = df.groupby('State')['Amount'].sum()
plt.figure(figsize=(12, 6))
plt.bar(state_amount.index, state_amount.values, color='lightseagreen')
plt.title('Total Amount by State')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

### **Orders by Marital Status and Gender**

Counts orders based on both marital status and gender.

Reveals how buying habits differ across marital and gender groups

```python
marital_status_orders = df.groupby(['Marital_Status', 'Gender'])['Orders'].sum().unstack()
marital_status_orders.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'lightpink'])
plt.title('Orders by Marital Status and Gender')
plt.xlabel('Marital Status')
plt.ylabel('Total Orders')
plt.show()
```

### **Distribution Of Genders**

Shows the number or proportion of males and females in the dataset.

Provides insights into the gender makeup of the population or sample being analyzed.

```python
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['mediumslateblue', 'darkturquoise'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
